# SMOTE TO HANDLE CLASS IMBALANCE:-
We use SMOTE to handle class imbalance by generating synthetic data for the minority class.

```python
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from imblearn.over_sampling import SMOTENC

# =========================================================================
# STEP 1: DATA LOADING AND PREPROCESSING
# =========================================================================
print("âœ… Step 1: Data Loading & Preprocessing")
try:
    data = pd.read_csv('dataset(Liver Disease).csv', delimiter=';')
except FileNotFoundError:
    print("Error: The file 'dataset(Liver Disease).csv' was not found.")
    exit()

data.replace(r'^\s*NA\s*$', np.nan, regex=True, inplace=True)
for col in data.columns:
    if col not in ['category', 'sex']:
        data[col] = pd.to_numeric(data[col], errors='coerce')

for col in data.select_dtypes(include=np.number).columns:
    data[col].fillna(data[col].mean(), inplace=True)

data = pd.get_dummies(data, columns=['sex'], drop_first=True)
X = data.drop('category', axis=1)
y = data['category']

# =========================================================================
# STEP 2: TRAIN-TEST SPLIT AND SCALING
# =========================================================================
print("âœ… Step 2: Splitting Data")
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# =========================================================================
# STEP 3: APPLY SMOTE TO HANDLE CLASS IMBALANCE
# Note: SMOTENC is used because your data has both numerical and categorical features.
# 'sex_m' is the categorical column in your data.
# =========================================================================
print("\nðŸ”„ Step 3: Applying SMOTE to the training data...")
print("Original class distribution (before SMOTE):\n", y_train.value_counts())

# Find the index of the categorical column in the scaled data
categorical_features_indices = [X.columns.get_loc('sex_m')]

# Set k_neighbors to be less than or equal to the smallest class size in y_train
# The smallest class 'suspect_disease' has 5 samples in y_train
smote_nc = SMOTENC(categorical_features=categorical_features_indices, random_state=42, k_neighbors=4)
X_train_res, y_train_res = smote_nc.fit_resample(X_train_scaled, y_train)

print("Resampled class distribution (after SMOTE):\n", y_train_res.value_counts())
print("âœ… SMOTE complete.\n")

# =========================================================================
# STEP 4: MODEL TRAINING ON RESAMPLED DATA
# Now train your models on the new, balanced dataset (X_train_res, y_train_res)
# =========================================================================
print("ðŸš€ Step 4: Training Models on Resampled Data")
models = {
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Naive Bayes": GaussianNB(),
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "SVM": SVC(kernel='rbf', random_state=42)
}

results_with_smote = {}
for name, model in models.items():
    print(f"  -> Training {name}...")
    model.fit(X_train_res, y_train_res)
    y_pred = model.predict(X_test_scaled)
    accuracy = accuracy_score(y_test, y_pred)
    results_with_smote[name] = {'accuracy': accuracy, 'predictions': y_pred}

print("\n")
# =========================================================================
# STEP 5: EVALUATE RESULTS
# =========================================================================
print("ðŸ“Š Step 5: Evaluation with SMOTE")
for name, res in results_with_smote.items():
    print(f"\n--- {name} Results (with SMOTE) ---")
    print(f"Accuracy: {res['accuracy']:.4f}")
    print("Classification Report:\n", classification_report(y_test, res['predictions']))
    cm = confusion_matrix(y_test, res['predictions'])
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
    plt.title(f'{name} Confusion Matrix (SMOTE)')
    plt.show()

print("\nâœ… All steps complete. You have successfully implemented SMOTE to handle class imbalance.")
```

# Hyperparameter Tuning for Random Forest

```python
from sklearn.model_selection import train_test_split, GridSearchCV
# Initial model training to find the best model (Random Forest)
models = {
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Naive Bayes": GaussianNB(),
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "SVM": SVC(kernel='rbf', random_state=42)
}
results = {}
for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    accuracy = accuracy_score(y_test, y_pred)
    results[name] = {'accuracy': accuracy, 'predictions': y_pred}

print("Random Forest (Initial Accuracy):", results["Random Forest"]['accuracy'])

# =========================================================================
# STEP 1: HYPERPARAMETER TUNING FOR RANDOM FOREST
# =========================================================================

print("\nðŸš€ Step 1: Hyperparameter Tuning for Random Forest")
rf_model_for_tuning = RandomForestClassifier(random_state=42)

# Parameters to test
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [10, 20, 30],
    'min_samples_leaf': [1, 2]
}

# Perform Grid Search
grid_search = GridSearchCV(estimator=rf_model_for_tuning, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train_scaled, y_train)

# Best model and its performance
best_rf_model = grid_search.best_estimator_
y_pred_tuned = best_rf_model.predict(X_test_scaled)
tuned_accuracy = accuracy_score(y_test, y_pred_tuned)

print(f"Best Parameters Found: {grid_search.best_params_}")
print(f"Random Forest (Tuned Accuracy): {tuned_accuracy:.4f}")
print("âœ… Tuning complete.\n")

# =========================================================================
# STEP 2: MODEL INTERPRETABILITY (FEATURE IMPORTANCE)
# =========================================================================

print("ðŸ” Step 2: Visualizing Feature Importance")
# Get feature importances from the best-tuned Random Forest model
feature_importances = best_rf_model.feature_importances_
features = X.columns
importance_df = pd.DataFrame({'Feature': features, 'Importance': feature_importances})

# Sort by importance and plot
importance_df = importance_df.sort_values(by='Importance', ascending=False)
plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=importance_df, palette='rocket')
plt.title('Feature Importance for Tuned Random Forest Model', fontsize=16)
plt.xlabel('Importance', fontsize=12)
plt.ylabel('Feature', fontsize=12)
plt.tight_layout()
plt.show()

print("âœ… All steps completed. You have a robust and well-explained project now.")
```

```python
from sklearn.model_selection import KFold, cross_val_score
box_plot_models = [
    ('Logistic Regression', LogisticRegression(max_iter=1000, random_state=42)),
    ('Random Forest', RandomForestClassifier(n_estimators=100, random_state=42)),
    ('SVM', SVC(kernel='rbf', random_state=42)),
    ('Naive Bayes', GaussianNB())
]

# Get the cross-validation scores for each model
results_cv = []
names_cv = []
scoring_metric = 'accuracy'
for name, model in box_plot_models:
    # Use cross-validation to get 10 accuracy scores for each model
    cv_results = cross_val_score(model, X_train_scaled, y_train, cv=KFold(n_splits=10, shuffle=True), scoring=scoring_metric)
    results_cv.append(cv_results)
    names_cv.append(name)
    print(f"{name}: {cv_results.mean():.4f} ({cv_results.std():.4f})")

# Create the box plot
plt.figure(figsize=(12, 8))
plt.boxplot(results_cv, labels=names_cv, showmeans=True)
plt.title('Model Accuracy Stability Comparison (Cross-Validation)', fontsize=16)
plt.xlabel('Model', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

models = {
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Naive Bayes": GaussianNB(),
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "SVM": SVC(kernel='rbf', random_state=42)
}
results = {}
for name, model in models.items():
    model.fit(X_train_scaled, y_train) # Note: Using unscaled X_train for simplicity here
    y_pred = model.predict(X_test_scaled)
    results[name] = {'predictions': y_pred}


# --- Step 2: Scatter Plot Visualization ---
print("ðŸ“Š Generating Scatter Plots with PCA...")

# Reduce features to 2 principal components for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_test_scaled)
pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
pca_df['True Label'] = y_test.reset_index(drop=True)

# Create subplots for each model's visualization
fig, axes = plt.subplots(2, 2, figsize=(16, 14))
fig.suptitle('Model Prediction Visualization using PCA (2D)', fontsize=20, y=1.02)
axes = axes.flatten()

# Plot the scatter plot for each model
for i, (name, res) in enumerate(results.items()):
    pca_df['Predicted Label'] = res['predictions']

    # Identify correct and incorrect predictions
    pca_df['Correct'] = pca_df['True Label'] == pca_df['Predicted Label']

    # Plotting correct predictions
    sns.scatterplot(
        x='PC1',
        y='PC2',
        data=pca_df[pca_df['Correct']],
        hue='True Label',
        style='Correct',
        markers={True: 'o', False: 'X'}, # Added False marker
        palette='tab10',
        ax=axes[i],
        s=100
    )

    # Plotting incorrect predictions
    sns.scatterplot(
        x='PC1',
        y='PC2',
        data=pca_df[~pca_df['Correct']],
        hue='Predicted Label',
        style='Correct',
        markers={True: 'o', False: 'X'}, # Added False marker
        palette='Set1',
        ax=axes[i],
        s=150
    )

    axes[i].set_title(f'{name} Predictions', fontsize=14)
    axes[i].set_xlabel('Principal Component 1', fontsize=12)
    axes[i].set_ylabel('Principal Component 2', fontsize=12)
    axes[i].legend(title='Labels', loc='best')

plt.tight_layout()
plt.show()

print("âœ… Scatter plots generated successfully.")
```
